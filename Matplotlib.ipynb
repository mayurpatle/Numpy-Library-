```python
plt.plot([2, 4 ,  10 , 7   , 56  , 57  ,58  ,32  , 98 ,35  , 77 , 78   ])
plt.ylabel("numbers")
plt.xlabel("indices")  
plt.title("MyPlot")
plt.grid() 

plt.show() 
```

numpy  and matplotlib together



```python
import  numpy as  np 

t  = np.arange(0. ,  5. , 0.2  )


plt.plot( t ,  t**2  ,  'b--'  , label = '^2')
plt.plot( t  , t **2.2  ,  'rs ' ,  label  = '^2.2')
plt.plot( t  , t**2.4 ,  'g^'  ,  label  =  '^2.4')
plt.grid() 
plt.legend () #helps to understand  which line peerforms what  -- extremeyl readable 


plt.show() 
```

```python
x  = [ 1 , 2 ,3  ,4 ]
y  =  [ 3  , 4 ,5  , 6  ]
plt.plot( x , y  ,  linewidth = 5.0 )
plt.show()  
```

Two plots simultaeniuly  


```python
a  =  [1  ,2  ,3  ,4 ] 
b = [ 1  , 4 ,  9 ,  16  ]
c  =  [ 1  , 2 , 3 ,4 ] 
d = [  2 , 4  , 6  , 8  ]

lines =  plt.plot(a , b  , c ,d )

plt.setp(lines[0]  ,  color  =  'r'  , linewidth  = 2.0)
plt.setp (lines[1]  , color =  'b'  , linewidth = 3.0)
plt.grid() 
plt.legend() 
plt.show() 
```

Subplots 



```python
def  func(t):
    return np.exp(-t) * np.cos(2  * np.pi* t  )

t1  =  np.arange(0.0 ,   5.0 ,  0.2  )
t2  =  np.arange(0.0   ,  5.0 ,  0.02 )


plt.figure(1)
plt.subplot(211)
plt.grid() 
plt.plot( t1 ,  func(t1)  ,  'b-')

plt.subplot(212)
plt.plot(t2  ,  np.cos(2  * np.pi *  t2)   ,  'r--')
plt .show() 


```
